```python
import random

def yazı_tura(num_flips):
  """
  Simulates a series of coin tosses and prints the results.

  Args:
      num_flips: The number of times to flip the coin.
  """
  # List of possible outcomes
  outcomes = ["Yazı", "Tura"]

  # Simulate the coin tosses
  for i in range(num_flips):
    choice = random.choice(outcomes)
    print(choice)

# Call the yazı_tura function with 100 flips
yazı_tura(100)
```
